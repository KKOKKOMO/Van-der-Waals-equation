```python
import matplotlib.pyplot as plt
import pandas as pd

# Пример загрузки данных из CSV (замените на свой файл или DataFrame)
df = pd.read_csv('Lennard-Jones_T')

# Для примера создадим фиктивный DataFrame с вашими колонками



plt.errorbar(
    df['T'],        # по X — температура
    df['P_mean'],   # по Y — давление
    xerr=df['P_std'],  # вертикальные ошибки (погрешность давления)
    fmt='o',
    ecolor='black',
    capsize=5
)


plt.plot(df['T'], df['P_ideal'], 'r--', label='Идеальный газ (P_ideal)')

plt.xlabel('Температура T')
plt.ylabel('Давление P')
plt.title('Зависимость давления от температуры с погрешностями')
plt.legend()
plt.grid(True)

# Раскомментируйте, если хотите попробовать логарифмическую шкалу
#plt.yscale('log')

plt.show()
```
